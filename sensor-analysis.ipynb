```python
import pandas as pd

df_inflow = pd.read_csv('/kaggle/input/wastewaterplant-data/dotok.txt', sep="   ", header=None, engine="python")
df_inflow.columns = ['t(d)', 'Q', 'T', 'COD', 'NH4', 'TN', 'TSS', 'TP']
df_inflow
```

```python
df_flow = pd.read_csv('/kaggle/input/wastewaterplant-data/pretok.txt', sep="   ", header=None, engine="python")
df_flow.columns = ['t(d)', 'Qintr', 'Qr', 'Qw', 'Quprimary', 'Quthickener', 'Qodewatering', 'Qmetal']
df_flow
```

```python
df_oxygen = pd.read_csv('/kaggle/input/wastewaterplant-data/kisik.txt', sep="   ", header=None, engine="python")
df_oxygen.columns = ['t(d)', 'DO5', 'DO6', 'DO6ref', 'DO7']
df_oxygen
```

```python
df_quality = pd.read_csv('/kaggle/input/wastewaterplant-data/kvaliteta.txt', sep="   ", header=None, engine="python")
df_quality.columns = ['t(d)', 'EQI']
df_quality
```

```python
df_cost = pd.read_csv('/kaggle/input/wastewaterplant-data/stroski.txt', sep="   ", header=None, engine="python")
df_cost.columns = ['t(d)', 'SC', 'AE', 'PE', 'ME', 'OCI']
df_cost
```

```python
df_inflow.set_index('t(d)', inplace=True)
df_flow.set_index('t(d)', inplace=True)
df_oxygen.set_index('t(d)', inplace=True)
df_quality.set_index('t(d)', inplace=True)
df_cost.set_index('t(d)', inplace=True)
```

```python
df = pd.concat([df_inflow, df_flow, df_oxygen, df_quality, df_cost], axis=1)
df
```

```python
y_EQI = df['EQI']
y_OCI = df['OCI']
X = df.drop(columns=['EQI', 'OCI', 'SC', 'AE', 'PE', 'ME'])
```

### Modeling EQI

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y_EQI, train_size=0.9, random_state=52)
```

```python
from sklearn.ensemble import RandomForestRegressor

rf_reg = RandomForestRegressor(n_estimators=100, bootstrap=True, random_state=42, oob_score=True)
rf_reg.fit(X_train, y_train)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
print(r2_score(y_train, rf_reg.predict(X_train)))
```

```python
rf_reg.oob_score_
```

```python
y_pred = rf_reg.predict(X_test)
```

```python
import numpy as np
from sklearn.metrics import r2_score, mean_squared_error

mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
print(r2_score(y_test, y_pred))
print(rmse)
```

```python
# Calculate feature importances
importances = rf_reg.feature_importances_

# Visualize Feature Importance
# Sort feature importances in descending order
indices = np.argsort(importances)[::-1]

# Rearrange feature names so they match the sorted feature importances
names = [X_train.columns[i] for i in indices]

plt.figure(figsize = (12, 5))
sns.set_style("whitegrid")
chart = sns.barplot(x = names, y=importances[indices])
plt.xticks(
    rotation=45, 
    horizontalalignment='right',
    fontweight='light'  
)
plt.title('Random forest. Feature importance')
plt.tight_layout()
```

```python
import lime
import lime.lime_tabular

explainer = lime.lime_tabular.LimeTabularExplainer(X_train.to_numpy(), feature_names=X_train.columns, class_names=['EQI'], verbose=True, mode='regression')
```

```python
plt.figure(figsize=(10,10))
plt.scatter(y_test, y_pred, c='crimson')
plt.yscale('log')
plt.xscale('log')

p1 = max(max(y_pred), max(y_test))
p2 = min(min(y_pred), min(y_test))
plt.plot([p1, p2], [p1, p2], 'b-')
plt.xlabel('True Values', fontsize=15)
plt.ylabel('Predictions', fontsize=15)
plt.axis('equal')
plt.show()
```

```python
i = 125
exp = explainer.explain_instance(X_test.to_numpy()[i], rf_reg.predict, num_features=20)
exp.show_in_notebook(show_table=True)
```

### Dataset 24 hour delay

Time is sampled in 0.01d = 14.4 minutes, every 4th row represents almost 1 hour of data.
Therefore, 24 hours represent 4 * 24 rows.

```python
df_24h_delay = df.copy(deep=True)
```

```python
df_24h_delay['EQI'] = df_24h_delay['EQI'].shift(-4*24)
df_24h_delay['OCI'] = df_24h_delay['OCI'].shift(-4*24)
```

```python
df_24h_delay.info()
```

```python
df_24h_delay.dropna(inplace=True)
```

```python
df_24h_delay.info()
```

```python
df_24h_delay.describe()
```

```python
y_EQI = df_24h_delay['EQI']
y_OCI = df_24h_delay['OCI']
X = df_24h_delay.drop(columns=['EQI', 'OCI', 'SC', 'AE', 'PE', 'ME'])
```

### Modelling EQI with delayed ds

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y_EQI, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor

rf_reg = RandomForestRegressor(n_estimators=250, bootstrap=True, random_state=42, oob_score=True)
rf_reg.fit(X_train, y_train)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
print(r2_score(y_train, rf_reg.predict(X_train)))
```

```python
rf_reg.oob_score_
```

```python
y_pred = rf_reg.predict(X_test)
```

```python
import numpy as np
from sklearn.metrics import r2_score, mean_squared_error

mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
print(r2_score(y_test, y_pred))
print(rmse)
```
